```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.style as style
style.use('fivethirtyeight')
```

```python
# Func to turn date legible.
def adjust_year(year):
    adjust = year.split('-');
    return str(adjust[1]+'-'+str(int(adjust[0])+2015) )

# Units solds by shampoo branding
shampoo = pd.read_csv('https://raw.githubusercontent.com/jbrownlee/Datasets/master/shampoo.csv');
shampoo['Date'] = pd.to_datetime(shampoo['Month'].apply(adjust_year), format='%m-%Y')
shampoo = shampoo[['Date', 'Sales']]
```

```python
plt.figure(figsize=(14, 7))
sns.lineplot(x='Date', y='Sales', data=shampoo)
plt.tick_params(axis='x', rotation=45)
```

```python
# Temperatures dataset
temperature = pd.read_csv('https://raw.githubusercontent.com/jbrownlee/Datasets/master/daily-min-temperatures.csv')
temperature['Date'] = pd.to_datetime(temperature['Date'])
```

```python
plt.figure(figsize=(14, 7))
sns.lineplot(x='Date', y='Temp', data=temperature)
plt.tick_params(axis='x', rotation=45)
```

```python
# Sunspot Dataset
sunspot = pd.read_csv('https://raw.githubusercontent.com/jbrownlee/Datasets/master/monthly-sunspots.csv')
sunspot['Date'] = pd.to_datetime(sunspot['Month']); sunspot= sunspot[['Sunspots', 'Date']]
```

```python
plt.figure(figsize=(14, 7))
sns.lineplot(x='Date', y='Sunspots', data=sunspot)
plt.tick_params(axis='x', rotation=45)
```

### Topic one. Stationary or non-Stationary.

#Lets apply two test Adfuller and kpss
Adfuller Ho is non-Stationary.

if > 0.05 is Non-Stationary and < 0.05 is Stationary.

```python
from statsmodels.tsa.stattools import adfuller
```

```python
shampooadfuller = adfuller(shampoo['Sales'].values)

print('Shampoo p value: ', str(shampooadfuller[1]), '. Timeseries is non-stationary')
```

```python
tempadfuller = adfuller(temperature['Temp'].values)

print('Temperature p value: ', str(tempadfuller[1]), '. Timeseries is stationary')
```

```python
sunspotsadfuller = adfuller(sunspot['Sunspots'].values)

print('Sunspots p value: ', str(sunspotsadfuller[1]), '. Timeseries is stationary')
```
