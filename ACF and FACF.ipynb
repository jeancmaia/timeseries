```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.style as style
from statsmodels.tsa.seasonal import seasonal_decompose
from pylab import rcParams
from matplotlib import pyplot
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
from statsmodels.tsa.stattools import pacf, acf
import matplotlib.pyplot as plt


style.use('fivethirtyeight')
```

```python
# Func to turn date legible.
def adjust_year(year):
    adjust = year.split('-');
    return str(adjust[1]+'-'+str(int(adjust[0])+2015) )

# Units solds by shampoo branding
shampoo = pd.read_csv('https://raw.githubusercontent.com/jbrownlee/Datasets/master/shampoo.csv');
shampoo['Date'] = pd.to_datetime(shampoo['Month'].apply(adjust_year), format='%m-%Y')
shampoo = shampoo[['Date', 'Sales']]
```

```python
acf_res = pd.DataFrame({})
acf_res['lags'] = [ i for i in range(0,26)]
acf_res['correlation'] = acf(shampoo['Sales'], nlags=25)
plt.figure(figsize=(14, 7))
sns.pointplot(x='lags', y='correlation', data=acf_res, color='green')
```

```python
pacf_res = pd.DataFrame({})
pacf_res['lags'] = [ i for i in range(0,26)]
pacf_res['correlation'] = pacf(shampoo['Sales'], nlags=25, method='ywmle')
plt.figure(figsize=(14, 7))
sns.pointplot(x='lags', y='correlation', data=pacf_res, color='green')
```
